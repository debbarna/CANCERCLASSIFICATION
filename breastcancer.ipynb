```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

importing of different libraries required for breast cancer classification

```python
#loading the data from sklearn
breast_cancer_datasets=sklearn.datasets.load_breast_cancer()
```

```python
#loading data to a data frame
data_frame=pd.DataFrame(breast_cancer_datasets.data,columns=breast_cancer_datasets.feature_names)
```

```python
# ADDING THE target COLUMN TO THE FRAME

data_frame['label']=breast_cancer_datasets.target
```

```python

X=data_frame.drop(columns='label',axis=1)
Y=data_frame['label']
```



> Add blockquote
SPILTING THE DATA INTO TRAINING DATA AND TESTING DATA


```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
scaler =StandardScaler()
X_train_std=scaler.fit_transform(X_train)
X_test_std=scaler.transform(X_test)
```

```python
print(X_train_std)
```

```python
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
#setting up the layers of my neural network

model=keras.Sequential([
    keras.layers.Flatten(input_shape=(30,)),
    keras.layers.Dense(20,activation='relu'),
    keras.layers.Dense(2,activation='sigmoid')
])
```

```python
#compiling the neural network
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
#training the neura network
history=model.fit(X_train_std,Y_train,validation_split=0.1,epochs=10)
```

```python
#visuaising accuracy and loss
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['training data','validation data'],loc='lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['training data','validation data'],loc='lower right')
```

```python
loss,accuracy=model.evaluate(X_test_std,Y_test)
print("test accuracy",accuracy)
```
